```python
def exist(board, word) -> bool:
    
    def all_index(copy_board, elem) -> list:
        return [index for index, value in enumerate(copy_board) if value == elem]
    
    
    def true_turn_index(numbers, index, board_width) -> list:
        true_turn = []
        all_turn = [index+board_width, 
                    index-board_width]
        
        if (index+1)%board_width != 0:
            all_turn.append(index+1)
            
        if (index//board_width) >= (index-1)//board_width:
            all_turn.append(index-1) 
        
        for i in all_turn:
            if i in numbers:
                true_turn.append(i)
        return true_turn
    
    
    def true_turn_letter(true_turn, board_list) -> list:
        true_letter = []
        for i in true_turn:
            if i != '*':
                true_letter.append(board_list[i])
        return true_letter
    
    board_list = list(''.join([''.join(i) for i in board]))
    board_width = len(board[0])
    board_length = len(board_list)
    
    
    
    up_word = word + '%'
    copy_board = board_list[:]
    used_numbers = []
    numbers = list(range(board_length))
    count_turn = 1
    data_board = []
    next_turn_in = []
    print(copy_board[:board_width])
    print(copy_board[board_width:board_width*2])
    print(copy_board[board_width*2:],'\n')
    for i_letter, letter in enumerate(up_word):
        if count_turn == 0:
            break
        count_turn = 0
        all_index_letter = all_index(copy_board, letter)
        if any(x in all_index_letter for x in next_turn_in) or not next_turn_in != []:
            for i in all_index_letter:
                print('='*50)
                next_true_number = true_turn_index(numbers,i,board_width)
                next_true_letter = true_turn_letter(next_true_number,copy_board)
                print(letter,f'({i})','->', up_word[i_letter+1], next_true_letter, next_true_number)
                next_letter = up_word[i_letter+1]
                if next_letter == '%':
                    return True
                if up_word[i_letter+1] in next_true_letter:
                    copy_board[i] = '*'
                    numbers[i] = '*'
                    next_turn_in = next_true_number[:]
                    print(copy_board[:board_width])
                    print(copy_board[board_width:board_width*2])
                    print(copy_board[board_width*2:],'\n')
                    count_turn += 1
                else:
                    print('NO NEXT TURN')
                
        else:
            continue
        print('*'*50,'NEW LETTER')

    return False

board, word = [["A","B","C","E"],["S","F","C","S"],["A","D","E","E"]], "ABCCED" #true
#board, word = [["A","B","C","E"],["S","F","C","S"],["A","D","E","E"]], "SEE" #true
#board, word = [["A","B","C","E"],["S","F","C","S"],["A","D","E","E"]], "ABCB" #false
#board, word = [["a","b"],["c","d"]],"abcd" #false
#board, word = [["a","b"],["c","d"]],"acdb" #true
#board, word = [["C","A","A"],["A","A","A"],["B","C","D"]],"AAB" #true

print(exist(board, word))
```
